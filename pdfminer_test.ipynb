```python
import os
import pdfminer
from pdfminer.image import ImageWriter
from pdfminer.high_level import extract_text, extract_pages

def get_image(layout_object):
    """Recursively locate Image objects in page_layout."""
    if isinstance(layout_object, pdfminer.layout.LTImage):
        return [layout_object]
    if isinstance(layout_object, pdfminer.layout.LTContainer):
        img_list = []
        for child in layout_object:
            img_list += get_image(child)
        return img_list
    else:
        return []

def extract_pdf_img_and_text(pdf_filepath, output_text_path):
    """Extract images and text from the PDF file."""
    # Extract text and write to file
    text = extract_text(pdf_filepath)
    with open(output_text_path, 'w', encoding='utf-8') as text_file:
        text_file.write(text)

    # Extract images
    iw = ImageWriter('output_dir')  # Make sure 'output_dir' exists
    for page_layout in extract_pages(pdf_filepath):
        image_list = get_image(page_layout)
        if image_list:
            for image in image_list:
                iw.export_image(image)

def find_pdf_path(pdf_base_dir, arxiv_id):
    """Recursively search for the PDF file corresponding to the given arxiv_id."""
    target_filename = arxiv_id + 'v1.pdf'
    for root, dirs, files in os.walk(pdf_base_dir):
        if target_filename in files:
            return os.path.join(root, target_filename)
    return None

def parse_pdfs(arxiv_ids):
    """Parse each PDF corresponding to an arxiv_id and save the text output and images."""
    for arxiv_id in arxiv_ids:
        # Find the path to the PDF file
        article_path = find_pdf_path(pdf_base_dir, arxiv_id)
        
        if article_path:
            # Define the output path for the text file
            output_text_path = os.path.join(output_base_dir, arxiv_id + '_output_file.txt')
            
            try:
                # Extract images and text from the PDF
                extract_pdf_img_and_text(article_path, output_text_path)
                
                print(f"Successfully processed {arxiv_id}")
            except Exception as e:
                print(f"Error processing {arxiv_id}: {e}")
        else:
            print(f"File not found for {arxiv_id}")

# Base directories
pdf_base_dir = '/home/jovyan/shared/notebooks/academic-rag-data/pdfupload/'
output_base_dir = '/home/jovyan/work/academic-rag/pdfminer_output/'

# Call the function with the arxiv_ids_2024 list
parse_pdfs(arxiv_ids_2024)
```
